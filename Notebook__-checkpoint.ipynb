# Comparing netMHCpan ranks to peptides with cystiene

Creating a a dist graphs for the cysteine and netMHCpan ranks

```python
import seaborn as sns

Cystine_df = df[df["cysteine"] == True]
cystine_rank1_array = np.array(Cystine_df[MHC_TYPES[0]])
```

```python
# Plot
sns.set_style("white")
Cystine_df = df[df["cysteine"] == True]
FCystine_df = df[df["cysteine"] == False]
for i in MHC_TYPES:
    cystine_rank1_array = np.array(Cystine_df[i])    
    fcystine_rank1_array = np.array(FCystine_df[i])
    kwargs = dict(hist_kws={'alpha':.6}, kde_kws={'linewidth':2})

    plt.figure(figsize=(10,7), dpi= 80)
    xs = sns.distplot(cystine_rank1_array, color="dodgerblue", label="cysteine true")
    sns.distplot(fcystine_rank1_array, color="orange", label="cysteine false")

    plt.xlabel(i.replace("_rank", "") + " ranks")
    plt.legend();
```

```python
amino_acids = list(set("".join([df["peptide"][i] for i in range(len(df["peptide"]))])))
```

```python
Chooped_df = df[df["Chopped"] == True]
NChopped_df = df[df["Chopped"] == False]
chopped_dict = {}
nchopped_dict = {}
for i in Chooped_df["peptide"]:
    last_letter = i[-1]
    if(last_letter not in chopped_dict):
        chopped_dict[last_letter] = 0
    chopped_dict[last_letter] += 1
    
for i in NChopped_df["peptide"]:
    last_letter = i[-1]
    if(last_letter not in nchopped_dict):
        nchopped_dict[last_letter] = 0
    nchopped_dict[last_letter] += 1


number_of_chooped = np.sum(list(chopped_dict.values()))
number_of_unChopped = np.sum(list(nchopped_dict.values()))
```

```python
labels = chopped_dict.keys()
chopped_counter = (np.divide(list(chopped_dict.values()), number_of_chooped) * 10000).astype(int) / 100
nchopped_counter = (np.divide(list(nchopped_dict.values()), number_of_unChopped) * 10000).astype(int) / 100

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
fig.set_figheight(9)
fig.set_figwidth(25)
fig.set_dpi(200)
rects1 = ax.bar(x - width/2, chopped_counter, width, label='chopped true')
rects2 = ax.bar(x + width/2, nchopped_counter, width, label="chopped false")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Predicted to be chopped by amino acids')
print(x)
print(labels)
ax.set_xtick(np.arange(len(labels)), labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()


plt.show()
```

# BoxPlot

```python
new_list = []
for mhc_type in MHC_TYPES:
    for i, row in df.iterrows():
        new_list.append([row[mhc_type], row["Chopped"], mhc_type])

cols = ["rank", "chopped", "mhc_type"]
box_df = pd.DataFrame(new_list, columns=cols)
```

```python
import seaborn as sns

labels = MHC_TYPES

sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")

sns.set(font_scale = 1.3)
print(box_df.head())
plt.figure(figsize = (30,10))

ax = sns.boxplot(x="mhc_type", y="rank", hue="chopped",
                 data=box_df, palette="Set3")
ax = sns.swarmplot(x="mhc_type", y="rank", hue="chopped",
                 data=box_df, palette="Set3")
plt.legend(loc="upper right", title="chopped")


```
