# Comparing netchop predictions to netMHCpan ranks

creating a list of all the mhc supertypes

```python
mhcpan_df = pd.read_csv("mhc_resualt.csv")
MHC_TYPES = list(set(mhcpan_df["mhc_type"]))
MHC_TYPES = [i + "_rank" for i in MHC_TYPES]
print(MHC_TYPES)
```

# Chopped - ranks BoxPlot

```python
new_list = []
for mhc_type in MHC_TYPES:
    for i, row in df.iterrows():
        new_list.append([row[mhc_type], row["Chopped"], mhc_type])

cols = ["rank", "chopped", "mhc_type"]
box_df = pd.DataFrame(new_list, columns=cols)
```

```python
import seaborn as sns

labels = MHC_TYPES

sns.set_theme(style="whitegrid")


sns.set(font_scale = 1.3)
print(box_df.head())
plt.figure(figsize = (30,10))

ax = sns.boxplot(x="mhc_type", y="rank", hue="chopped",
                 data=box_df, palette="Set3")
ax = sns.stripplot(x="mhc_type", y="rank", hue="chopped", data=box_df, ax=ax, palette="Set2", split=True, jitter=0.2, alpha=0.6)
plt.legend(loc="upper right", title="chopped")

plt.savefig("plots/chopped-rank-boxplot", dpi=500, bbox_inches='tight')
```

# comaring amino acids - chopped 

```python
amino_acids = list(set("".join([df["peptide"][i] for i in range(len(df["peptide"]))])))
```

```python
Chooped_df = df[df["Chopped"] == True]
NChopped_df = df[df["Chopped"] == False]
chopped_dict = {}
nchopped_dict = {}
for i in Chooped_df["peptide"]:
    last_letter = i[-1]
    if(last_letter not in chopped_dict):
        chopped_dict[last_letter] = 0
    chopped_dict[last_letter] += 1
    
for i in NChopped_df["peptide"]:
    last_letter = i[-1]
    if(last_letter not in nchopped_dict):
        nchopped_dict[last_letter] = 0
    nchopped_dict[last_letter] += 1


number_of_chooped = np.sum(list(chopped_dict.values()))
number_of_unChopped = np.sum(list(nchopped_dict.values()))
```

```python
labels = chopped_dict.keys()
chopped_counter = (np.divide(list(chopped_dict.values()), number_of_chooped) * 10000).astype(int) / 100
nchopped_counter = (np.divide(list(nchopped_dict.values()), number_of_unChopped) * 10000).astype(int) / 100

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
fig.set_figheight(9)
fig.set_figwidth(25)
fig.set_dpi(200)
rects1 = ax.bar(x - width/2, chopped_counter, width, label='chopped true')
rects2 = ax.bar(x + width/2, nchopped_counter, width, label="chopped false")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Predicted to be chopped by amino acids')
print(x)
print(labels)
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()
plt.xlabel("amino acids")
plt.ylabel("amount")
plt.savefig("plots/aminoacids-chopped-amount.png", dpi=500, bbox_inches='tight')
plt.show()
```

# Entropy stuff

```python
from scipy.interpolate import make_interp_spline
entropy_df = pd.read_excel("entropy-vector.xlsx")

def calculate_entropy(pos):
    return float(entropy_df[entropy_df["Unnamed: 0"] == pos]["entropy"])
```

dist graph for entropy

```python
x = []
y = []
x2 = []
y2 = []
chopped_df8 = df8[df8["Chopped"] == True]
nchopped_df8 = df8[df8["Chopped"] == False]


for i, row in chopped_df8.iterrows():
    y.append(calculate_entropy(row["Unnamed: 0"] + 1 ))
    x.append(row["Unnamed: 0"])

for i, row in nchopped_df8.iterrows():
    y2.append(calculate_entropy(row["Unnamed: 0"] + 1))
    x2.append(row["Unnamed: 0"])

plt.xlabel("position")
plt.ylabel("entropy")

x = np.array(x)
y = np.array(y)
X_Y_Spline = make_interp_spline(x, y)
 
# Returns evenly spaced numbers
# over a specified interval.

X_ = np.linspace(x.min(), x.max(), 500)

Y_ = X_Y_Spline(X_)
 
# Plotting the Graph
plt.scatter(x, y, label="chopped" , alpha=0.3)


x2 = np.array(x2)
y2 = np.array(y2)
X_Y_Spline2 = make_interp_spline(x2, y2)
 
# Returns evenly spaced numbers
# over a specified interval.

X_2 = np.linspace(x2.min(), x2.max(), 500)

Y_2 = X_Y_Spline2(X_2)
 
# Plotting the Graph
plt.scatter(x2, y2, label="not chopped", alpha=0.3)

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.savefig("plots/entropy-position-chopped.png", dpi=500, bbox_inches='tight')
```
